Exploratory Data Analysis
<p>We explore the data to define the trend in GDP growth over the years</p>

```python
sum_columns=df[1:].sum()
print(sum_columns)
```

```python
# import other libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# The GDP trend in my country Kenya
plt.plot(df["Year"],df["Kenya"])
plt.title("GDP trend in Kenya")
plt.xlabel("Year")
plt.ylabel("GDP count")
plt.figure(figsize=(8,7.5))
```

The Kenyan Ecocnomy has significantly gronw over the years and the GDP is relatively high in the recent years .

```python
# demonstrating the GDP of EA countries
# listing down the columnc that represent the countries
columns=df[["Kenya", "Burundi", "Uganda", "Tanzania", "Rwanda"]]
plt.figure(figsize=(12,10))

for country in columns:
    plt.plot(df["Year"], df[country], label=country)
plt.xlabel("Year")
plt.ylabel("GDP Value")
plt.title("GDP for East Africa")
plt.legend(bbox_to_anchor=(1.15, 1.0), loc="upper left", borderaxespad=0)

plt.grid()
plt.show()
    
```

```python
# Checking the GDP growth for all the countries
cols=df.columns.tolist()[1:]
plt.figure(figsize=(8,8))
for country in cols:
    plt.plot(df["Year"], df[country], label=country)
plt.xlabel("Year")
plt.ylabel("GDP Value")
plt.title("GDP for Countries")
plt.legend(bbox_to_anchor=(1.15, 1.0), loc="upper left", borderaxespad=0)

plt.grid()
plt.show()
    
```

```python
togo_libya=df[["Togo", "Libya"]]
for country in togo_libya:
    plt.plot(df["Year"], df[country], label=country)
plt.title("GDP Difference between Togo and Libya")
plt.xlabel("Year")
plt.ylabel("GDP Value")
plt.legend(bbox_to_anchor=(1.15, 1.0), loc="upper left", borderaxespad=0)

plt.grid()
plt.show()
```

***Libya's GDP***-- Highest GDP Value
<p>Libya has had an unstable GDP over the years with significant high GDP growth in the years 1970-1980 attributed to huge oil exports, then the economy was significantly unstable from 1980s to 2000s. The instability could be as an impact of the civil war and political instability in the 2000s until currently. However, Libya has recorded the highest GDP value of over 9 billion US Dollars. 
Highest GDP in the whole of Africa.</p>
